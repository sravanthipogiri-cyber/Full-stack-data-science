```python
# First, install the required package
!pip install pyttsx3

# Then run your original code
import pyttsx3
from IPython.display import display, Audio
import os

# Initialize the text-to-speech engine
engine = pyttsx3.init()

# Get available voices
voices = engine.getProperty('voices')

# Set male voice
# voices[0] is usually male voice in most systems
engine.setProperty('voice', voices[0].id)

# Set the text to be spoken
text = "Welcome to Data Science"

# Set the output file path
output_file = 'text_to_speech_male.wav'

# Save to file
engine.save_to_file(text, output_file)
engine.runAndWait()

# Display audio in notebook
display(Audio(output_file, autoplay=False))
```
