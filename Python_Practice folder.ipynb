```python
# Write a program that checks if the input number is a prime number
num = 31
if num > 0:
    for i in range(2, num):
        if (num % i == 0):
            print("Not a PrimeNumber")
            break
    else:  # This else belongs to the for loop, not the if statement
        print("PrimeNumber")
else:
    print("Not a PrimeNumber")
```
